```python
from fastai.vision.all import *
```

```python
df_org = pd.read_csv('../csv/train_val_original.csv')
df_preprocssed = pd.read_csv('../csv/train_val_noise.csv')
```

```python
org_img_list = df_org.image.tolist()
processed_img_list = df_preprocssed.image.tolist()
```

```python
len(org_img_list) - len(processed_img_list)
```

```python
img_fn_list = set(org_img_list) - set(processed_img_list)
```

```python
mask_fn_list = [fn.replace('image', 'mask')for fn in img_fn_list]
```

```python
def get_imgs(img_list, mask_list): 
    pil_img_list = []
    pil_mask_list = []
    
    for im, mask in list(zip(img_list, mask_list)):        
        im = Image.open(im)
        pil_img_list.append(im)
        
        mask = Image.open(mask)
        pil_mask_list.append(mask)
        
    return pil_img_list, pil_mask_list
```

```python
img_list, mask_list = get_imgs(img_fn_list, mask_fn_list)
```

```python
count = 0
```

```python
import numpy as np
import matplotlib.pyplot as plt

w = 10
h = 10
fig = plt.figure(figsize=(20, 20))
columns = 5
rows = 5
for i in range(1, columns*rows +1):
    img = img_list[count + i-1]
    mask = mask_list[count + i-1]
    
    fig.add_subplot(rows, columns, i)
    plt.tight_layout()
    plt.imshow(img, cmap='gray')
    plt.imshow(mask, cmap='jet', alpha=0.4)
```
